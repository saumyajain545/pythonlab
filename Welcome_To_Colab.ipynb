```
# Import matplotlib library
import matplotlib.pyplot as plt

# Data
months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct']
sales = [120,150,170,160,180,200,210,190,220,230]

# Plot line chart
plt.plot(months, sales)

# Labels and title
plt.title("Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Sales")

# Show graph
plt.show()
```
