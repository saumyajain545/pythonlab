```python
# Import pandas library
import pandas as pd

# Create two Series with different indexes
s1 = pd.Series([10, 20, 30], index=['A', 'B', 'C'])
s2 = pd.Series([5, 15, 25], index=['B', 'C', 'D'])

# -------------------------------
# 1. Add both Series (s1 + s2)
# -------------------------------
print("Addition of Series (s1 + s2):")
result = s1 + s2
print(result)
print("===============================================")

# -------------------------------
# 2. Observe NaN values
# -------------------------------
print("NaN values appear for unmatched indexes (A and D)")
print(result)
print("===================================================")

# -------------------------------
# 3. Replace NaN with 0 and recompute
# -------------------------------
print("After replacing NaN with 0:")

# Use add() with fill_value = 0
result2 = s1.add(s2, fill_value=0)
print(result2)
```
