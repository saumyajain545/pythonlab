```python
# Import pandas library
import pandas as pd

# Create data
marks = [45, 67, 89, 34, 90, 55, 72]
subjects = ['Math', 'Physics', 'Chemistry', 'English', 'CS', 'Biology', 'History']

# Create a Series with subjects as index
s = pd.Series(marks, index=subjects)

# -------------------------------
# 1. Find subjects with marks > 70
# -------------------------------
print("Subjects with marks greater than 70:")
print(s[s > 70])
print("===========================================")
# -------------------------------
# 2. Show "Fail" for marks < 40 (only in output)
# -------------------------------
print("Marks after applying Fail condition:")
print(s.where(s >= 40, "Fail"))
print("===========================================")
# -------------------------------
# 3. Count number of subjects above average
# -------------------------------
avg = round(s.mean(), 2) # Calculate average marks
count = (s > avg).sum()  # Count values greater than average

print("Average Marks:", avg)
print("Subjects scoring above average:", count)
```
