```python
# Import pandas library
import pandas as pd

# Create dataset
data = [5, None, 15, 25, None, 35, 45]

# Create Series
s = pd.Series(data)

print("Original Data:")
print(s)
print("==========================================")

# -------------------------------
# 1. Fill missing values with median
# -------------------------------
median_val = s.median()      # Calculate median
s = s.fillna(median_val)     # Fill missing values

print("After filling missing values with median:")
print(s)
print("===============================================")

# -------------------------------
# 2. Filter values greater than 20
# -------------------------------
filtered = s[s > 20]

print("Values greater than 20:")
print(filtered)
print("=========================================")

# -------------------------------
# 3. Sort in descending order
# -------------------------------
sorted_s = filtered.sort_values(ascending=False)

print("Sorted in descending order:")
print(sorted_s)
print("=========================================")

# -------------------------------
# 4. Reset index
# -------------------------------
final = sorted_s.reset_index(drop=True)

print("After resetting index:")
print(final)
```
