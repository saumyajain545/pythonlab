```python
# Import pandas library
import pandas as pd

# Create dataset with missing values
data = [10, None, 30, None, 50, 60, None]

# Create Series
s = pd.Series(data)

# -------------------------------
# Fill missing values with mean
# -------------------------------
print("Fill missing values with mean:")
mean_val = s.mean()         # Calculate mean
s = s.fillna(mean_val)      # Fill missing values
print(s)
print("===================================")

# -------------------------------
# Count missing values before and after filling
# -------------------------------
print("Count missing values before and after filling:")

# Recreate original series to count before
s_original = pd.Series(data)

before = s_original.isnull().sum()
after = s.isnull().sum()

print("Before:", before)
print("After:", after)
print("===================================")

# -------------------------------
# Drop remaining nulls (if any)
# -------------------------------
print("Drop remaining nulls (if any):")
s = s.dropna()

print(s)
```
