```python
# Import pandas library
import pandas as pd

# Create dataset
scores = [88, 92, 79, 93, 85]
players = ['P1','P2','P3','P4','P5']

# Create Series
s = pd.Series(scores, index=players)

# -------------------------------
# 1. Rank players based on scores
# -------------------------------
print("Player Rankings:")
ranks = s.rank(ascending=False)   # Highest score gets rank 1
print(ranks)
print("======================================================")

# -------------------------------
# 2. Display top 3 players
# -------------------------------
print("Top 3 Players:")
top3 = s.sort_values(ascending=False).head(3)
print(top3)
print("========================================================")

# -------------------------------
# 3. Find player with lowest score
# -------------------------------
print("Player with lowest score:")
lowest = s.idxmin()   # Returns index (player name) of minimum score
print(lowest)
```
