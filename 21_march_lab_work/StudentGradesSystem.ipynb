```python
# Import pandas library
import pandas as pd

# Create dataset
marks = [95, 82, 67, 58, 73, 49, 88]
students = ['S1','S2','S3','S4','S5','S6','S7']

# Create Series
s = pd.Series(marks, index=students)

# -------------------------------
# Assign grades using apply()
# -------------------------------
print("Assigned Grades:")

# Define function for grading
def grade(x):
    if x >= 90:
        return 'A'
    elif x >= 75:
        return 'B'
    elif x >= 60:
        return 'C'
    else:
        return 'D'

# Apply function to Series
grades = s.apply(grade)

print(grades)
print("=================================================")

# -------------------------------
# Count number of students in each grade
# -------------------------------
print("Number of students in each grade:")

count = grades.value_counts()   # Count frequency of each grade
print(count)
```
