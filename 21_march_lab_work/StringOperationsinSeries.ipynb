```python
# Import pandas library
import pandas as pd

# Create dataset
names = ['anuj', 'rahul', 'sneha', 'kiran', 'amit']

# Create Series
s = pd.Series(names)

print("Original Names:")
print(s)
print("================================================")

# -------------------------------
# 1. Convert all names to uppercase
# -------------------------------
print("Names in Uppercase:")
upper_names = s.str.upper()   # Convert all strings to uppercase
print(upper_names)
print("================================================")

# -------------------------------
# 2. Find names containing letter 'a'
# -------------------------------
print("Names containing letter 'a':")
contains_a = s[s.str.contains('a')]   # Filter names having 'a'
print(contains_a)
print("==================================================")

# -------------------------------
# 3. Replace 'anuj' with 'Anuj Kumar'
# -------------------------------
print("After replacing 'anuj' with 'Anuj Kumar':")
replaced = s.replace('anuj', 'Anuj Kumar')   # Replace specific value
print(replaced)
```
