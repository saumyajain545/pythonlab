```python
# Import pandas
import pandas as pd

# Dataset (use float values to avoid warning)
salary = [25000.0, 32000.0, 28000.0, 40000.0, 50000.0]
emp = ['A', 'B', 'C', 'D', 'E']

# Create Series
s = pd.Series(salary, index=emp)

print("Original Salaries:")
print(s)
print()

# -------------------------------
# Increase salary by 10% where salary < 30000
# -------------------------------
s[s < 30000] = s[s < 30000] * 1.10

# -------------------------------
# Decrease salary by 5% where salary > 45000
# -------------------------------
s[s > 45000] = s[s > 45000] * 0.95

print("Updated Salaries:")
print(s)
print()

# -------------------------------
# Total salary expense
# -------------------------------
total = s.sum()
print("Total Salary Expense:", round(total, 2))
```
