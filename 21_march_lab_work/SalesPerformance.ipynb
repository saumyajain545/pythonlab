```python
# Import pandas library
import pandas as pd

# Create dataset
sales = [200, 450, 300, 150, 500, 700, 100]
days = ['Mon','Tue','Wed','Thu','Fri','Sat','Sun']

# Create Series
s = pd.Series(sales, index=days)

print("Original Sales:")
print(s)
print("===========================================")

# -------------------------------
# 1. Find best and worst sales day
# -------------------------------
best_day = s.idxmax()    # Day with highest sales
worst_day = s.idxmin()   # Day with lowest sales

print("Best Sales Day:", best_day)
print("Worst Sales Day:", worst_day)
print("============================================")

# -------------------------------
# 2. Calculate total and average sales
# -------------------------------
total = s.sum()
avg = s.mean()

print("Total Sales:", total)
print("Average Sales:", round(avg, 2))
print("=========================================")

# -------------------------------
# 3. Apply 20% bonus where sales > 400
# -------------------------------
bonus_sales = s.copy()
bonus_sales[bonus_sales > 400] = bonus_sales[bonus_sales > 400] * 1.20

print("Sales after applying 20% bonus:")
print(bonus_sales)
print("============================================")

# -------------------------------
# 4. Sort sales in descending order
# -------------------------------
sorted_sales = s.sort_values(ascending=False)

print("Sales sorted in descending order:")
print(sorted_sales)
```
