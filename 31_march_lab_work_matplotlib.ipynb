```
# Import required library
import matplotlib.pyplot as plt

# Input data
x = [1, 2, 3, 4, 5]  # x-axis values
y = [2, 4, 6, 8, 10] # y-axis values

# Create a line plot
plt.plot(x, y, color='blue', marker='o', linestyle='-', linewidth=2, markersize=6)

# Add title and labels
plt.title("Line Plot of x vs y")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Show grid
plt.grid(True)

# Display the plot
plt.show()
```

```
# Import required library
import matplotlib.pyplot as plt

# Input data
products = ['Product A', 'Product B', 'Product C', 'Product D', 'Product E']
sales = [120, 150, 90, 180, 75]

# Create a bar chart
plt.bar(products, sales, color='green', edgecolor='black')

# Add title and labels
plt.title("Sales of Products")
plt.xlabel("Products")
plt.ylabel("Sales (in units)")

# Show grid
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the plot
plt.show()
```

```
# Import required library
import matplotlib.pyplot as plt

# Input data
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [40, 25, 20, 15]

# Create a pie chart
plt.pie(values, labels=categories, autopct='%1.1f%%', startangle=90, colors=['gold', 'lightblue', 'lightgreen', 'pink'], explode=(0.1, 0, 0, 0))

# Add title
plt.title("Percentage Distribution of Categories")

# Display the plot
plt.show()
```

```
# Import required libraries
import matplotlib.pyplot as plt
import numpy as np

# Generate random dataset
data = np.random.randint(1, 101, 50)  # 50 random integers between 1 and 100

# Create a histogram
plt.hist(data, bins=10, color='skyblue', edgecolor='black')  # 10 bins

# Add title and labels
plt.title("Histogram of Random Dataset")
plt.xlabel("Value Ranges")
plt.ylabel("Frequency")

# Show grid
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the plot
plt.show()
```

```
# Import library
import matplotlib.pyplot as plt

# Input datasets
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]   # Dataset 1
y2 = [1, 3, 5, 7, 9]    # Dataset 2

# Plot first dataset
plt.plot(x, y1, color='blue', marker='o', linestyle='-', linewidth=2, label='Dataset 1')

# Plot second dataset
plt.plot(x, y2, color='red', marker='s', linestyle='--', linewidth=2, label='Dataset 2')

# Add title and labels
plt.title("Comparison of Two Datasets")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Add legend
plt.legend()

# Show grid
plt.grid(True)

# Display the plot
plt.show()
```

```
# Import library
import matplotlib.pyplot as plt
import numpy as np

# Input data points
x = [1, 2, 3, 4, 5]
y = [10, 15, 8, 20, 12]

# Create scatter plot
plt.scatter(x, y, color='blue', label='Data Points')

# Find index of maximum y-value
max_index = np.argmax(y)
max_x = x[max_index]
max_y = y[max_index]

# Highlight the maximum point
plt.scatter(max_x, max_y, color='red', s=100, label='Maximum Point')  # s=100 → size of marker

# Annotate the maximum point
plt.text(max_x + 0.1, max_y, f'Max: ({max_x},{max_y})', fontsize=10, color='red')

# Add title and labels
plt.title("Scatter Plot with Maximum Point Highlighted")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Add legend
plt.legend()

# Show grid
plt.grid(True)

# Display the plot
plt.show()
```

```
# Import library
import matplotlib.pyplot as plt

# Input data: Monthly revenue (in thousands)
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
revenue = [50, 60, 55, 70, 65, 80]

# Create a line plot
plt.plot(months, revenue, color='purple', marker='o', linestyle='-', linewidth=2, markersize=8)

# Add title and axis labels
plt.title("Monthly Revenue (Jan-Jun)", fontsize=14, fontweight='bold', color='darkblue')
plt.xlabel("Months", fontsize=12)
plt.ylabel("Revenue (in $1000s)", fontsize=12)

# Add grid
plt.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)

# Display the plot
plt.show()
```
