```
#to import pandas library
import pandas as pd

#----------------------------------------
data = [80, 90, 70, 45, 65]

#Indexing
subname = ['Science', 'Maths', 'Hindi', 'English', 'Computer']

#to create a series
marks = pd.Series(data, index=subname)

#----------------------------------------
#to display the series
print(marks)
```

```
#creating a series with Dictionary
product_info = pd.Series({'Keyboard': 560, 'Mouse': 700, 'Monitor': 500, 'Tablet': 800, 'Printer': 500})

#to display series
print("Product with Available quantity : ")
print(product_info)
print("--------------------------------------------------")
print("Quantity of Mouse : ", product_info['Mouse'])
```

```
#operations on Series
#to increment quantity of each product by 200
product_info = product_info + 200

print("Product with Available quantity : ")
print(product_info)
#
```

```
# display the product having quantity greater than 800

print("Product with quantity > 800: ")
print(product_info[product_info > 800])
```

```
# creating s series

data = pd.Series([50,60,80,50,72,60,1,3,4,1,2,3,4,60,8,9,60,80,90,50,70,1,3,5,80,7,0,5,0,6,0,80,52,62,55,63,70,50,30,61])

#to display first 5 values
print("First 5 values :")
print(data.head(5))

#to display last 5 values
print("Last 5 vallues :")
print(data.tail(5))

#to display count
print("count of each value are :")
print(data.value_counts())
```

```
# to display unique values
print("Unique values :")
print(data.unique())
# total count of unique values
print("-----------------------")
print("Total unique values :",data.nunique())
```

```
#creating a series
data = pd.Series([78,90,45,90,79],index=['Hindi','science','English','Maths','Computer'])
# to sort the data as per values
print("Sorted data as per values")
print(data.sort_values())
# to sort data as per index
print("--------------------------------")
print("Sorted data as per index:")
print(data.sort_index())
```

```
data = pd.Series([50,60,None,90,50,None])

print(data.isnull())
```

```
import numpy as np
data1 = pd.Series([50,60,None,90,30,np.nan,90,60,None])
# to fill empty values with 0
print(data1.fillna(0))
```

```
data2 = pd.Series([50,60,None,60,90,None,20,None])
#missing values treatement by filling data as average value
print(data2.fillna(data2.mean()))
```

```
data3 = pd.Series([60,90,None,50,34,None,None,78])
#dropping missing values
print(data3.dropna())
```
