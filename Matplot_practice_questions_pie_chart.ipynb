```
import matplotlib.pyplot as plt

brands = ["Samsung","Apple","Xiaomi","Oppo","Others"]
share = [30,25,20,15,10]

plt.pie(share, labels=brands, autopct='%1.1f%%')

plt.title("Mobile Market Share")
plt.show()
```

```
import matplotlib.pyplot as plt

# Data for browser usage
browsers = ["Chrome", "Firefox", "Edge", "Safari", "Others"]
usage = [60, 15, 10, 10, 5]

# Create pie chart
plt.pie(usage, labels=browsers, autopct='%1.1f%%')

# Title
plt.title("Browser Usage Share")

# Show plot
plt.show()
```
