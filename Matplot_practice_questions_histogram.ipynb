```
import matplotlib.pyplot as plt

scores = [45,50,55,60,62,65,68,70,72,75,78,80,83,85,90]

plt.hist(scores, bins=5)

plt.title("Student Exam Scores")
plt.xlabel("Score")
plt.ylabel("Frequency")

plt.show()
```
