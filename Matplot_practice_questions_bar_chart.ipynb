```
# Import library
import matplotlib.pyplot as plt

# Data
courses = ['Python', 'Java', 'C++', 'Web', 'AI']
students = [40, 35, 25, 30, 20]

# Create bar chart
plt.bar(courses, students)

# Title and labels
plt.title("Students in Different Courses")
plt.xlabel("Courses")
plt.ylabel("Number of Students")

# Show chart
plt.show()
```
